```python
# read overall file
femaleFigureSkater = pd.read_csv("femaleFigureSkater_text_overall.csv")

#total number of articles
femaleFigureSkater_total_articles = len(glob.glob('../femaleFigureSkater/new_csvs/*'))

# all nouns
femaleFigureSkater_noun= femaleFigureSkater[femaleFigureSkater.pos =="名詞"].sort_values("total", ascending=False)
femaleFigureSkater_noun["per_article"]= round(femaleFigureSkater_noun["articles"]/ femaleFigureSkater_total_articles,2)
femaleFigureSkater_noun["gender"]= "female"
femaleFigureSkater_noun["occupation"]= "figure skater"
femaleFigureSkater_noun.to_csv("femaleFigureSkater_noun.csv", index= False)
# print_result = pd.read_csv("femaleFigureSkater_noun.csv")

# top20 nouns
femaleFigureSkater_top20_noun= femaleFigureSkater_noun.head(20)
femaleFigureSkater_top20_noun.to_csv("femaleFigureSkater_top20_noun.csv", index= False)
# print_result = pd.read_csv("femaleFigureSkater_top20_noun.csv")

# all verbs
femaleFigureSkater_verb= femaleFigureSkater[femaleFigureSkater.pos =="動詞"].sort_values("total", ascending=False)
femaleFigureSkater_verb["per_article"]= round(femaleFigureSkater_verb["articles"]/ femaleFigureSkater_total_articles,2)
femaleFigureSkater_verb["gender"]= "female"
femaleFigureSkater_verb["occupation"]= "figure skater"
femaleFigureSkater_verb.to_csv("femaleFigureSkater_verb.csv", index= False)
# print_result = pd.read_csv("femaleFigureSkater_verb.csv")

# top20 verbs
femaleFigureSkater_top20_verb= femaleFigureSkater_verb.head(20)
femaleFigureSkater_top20_verb.to_csv("femaleFigureSkater_top20_verb.csv", index= False)
# print_result = pd.read_csv("femaleFigureSkater_top20_verb.csv")

# all adjs
femaleFigureSkater_adj= femaleFigureSkater[femaleFigureSkater.pos =="形容詞"].sort_values("total", ascending=False)
femaleFigureSkater_adj["per_article"]= round(femaleFigureSkater_adj["articles"]/ femaleFigureSkater_total_articles,2)
femaleFigureSkater_adj["gender"]= "female"
femaleFigureSkater_adj["occupation"]= "figure skater"
femaleFigureSkater_adj.to_csv("femaleFigureSkater_adj.csv", index= False)
# print_result = pd.read_csv("femaleFigureSkater_adj.csv")

# top20 adj
femaleFigureSkater_top20_adj= femaleFigureSkater_adj.head(20)
femaleFigureSkater_top20_adj.to_csv("femaleFigureSkater_top20_adj.csv", index= False)
```

```python
# merge nouns + verbs + adjs
femaleFigureSkater_top_words = pd.concat([femaleFigureSkater_noun, femaleFigureSkater_verb,femaleFigureSkater_adj], axis=0)
femaleFigureSkater_top_words.to_csv("femaleFigureSkater_top_words.csv", index= False)
print_result = pd.read_csv("femaleFigureSkater_top_words.csv")
print_result
```
