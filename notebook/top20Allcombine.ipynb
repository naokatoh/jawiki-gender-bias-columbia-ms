```python
import pandas as pd
```

```python
female = pd.read_csv("top20_all_female_only.csv")
male = pd.read_csv("top20_all_male_only.csv",encoding="SHIFT-JIS")
```

```python
female = female.add_prefix('female_')
female = female.rename(columns={'female_word': 'word'})
female = female.rename(columns={'female_occupation': 'occupation'})
female.head()
```

```python
male = male.add_prefix('male_')
male = male.rename(columns={'male_word': 'word'})
male = male.rename(columns={'male_occupation': 'occupation'})
male.head(9)
```

```python
male.info()
```

```python
test_merged = pd.merge(female, male, on="word", how ="outer")
test_merged
```

```python
top20_merged = pd.merge(female, male, how ="outer", on = ["word","occupation"])
top20_merged
```

```python
top20_merged.to_csv("top20_merged.csv", index=False)
```
